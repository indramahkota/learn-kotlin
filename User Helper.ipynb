```kotlin
data class UserEntity(
    var id: Long = 0,
    val userId: Long,
    val avatarUrl: String?,
    val name: String?,
    val login: String?,
    val bio: String?,
    val followers: Int?,
    val following: Int?,
    val location: String?,
    val email: String?
)

data class UserDto(
    val id: Long,
    val avatarUrl: String?,
    val name: String?,
    val login: String?,
    val bio: String?,
    val followers: Int?,
    val following: Int?,
    val location: String?,
    val email: String?
)

/* User DTO to Entity */
fun UserDto.toUserEntity() = UserEntity(
    userId = id,
    avatarUrl = avatarUrl,
    name = name,
    login = login,
    bio = bio,
    followers = followers,
    following = following,
    location = location,
    email = email
)

fun List<UserDto>.toListUserEntity() =
    this.map {
        it.toUserEntity()
    }
```
